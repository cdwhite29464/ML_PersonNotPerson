```python
import os
import matplotlib
import pandas as pd
from pathlib import Path
from fastai.vision import ImageDataBunch, get_transforms, imagenet_stats
from arcgis.gis import GIS
from arcgis.learn import FeatureClassifier
```

# Photos to be Sorted

```python
folder_dir = r"M:\GIS\3_Resources\3_7_Scripts\Python\ML_ImageClassification_PersonNotPerson\Example_Photos - Copy"
os.chdir(folder_dir)
os.makedirs("Person")
os.makedirs("Not Person")
```

# Call Data for Model

```python
nameCSV = "PnotPTPhotos.csv"
folderLoc = r"M:\GIS\3_Resources\3_7_Scripts\Python\ML_ImageClassification_PersonNotPerson\TrainingData"

new_train_df = pd.read_csv(os.path.join(folderLoc, nameCSV))

Path_df=Path(folderLoc)                                           # path to the downloaded data
data = ImageDataBunch.from_df(path=Path_df,
                              folder='',
                              df=new_train_df,
                              fn_col=0,
                              label_col=1,
                              valid_pct=0.2,
                              seed=42,
                              bs=16,
                              size=224,
                              num_workers=2).normalize(imagenet_stats)
```

# Call Model

```python
modelLocation = r"M:\GIS\3_Resources\3_7_Scripts\Python\ML_ImageClassification_PersonNotPerson\TrainingData\PNP_ModelFull\PNP_ModelFull.emd"
```

# Call FeatureClassifier Object with Model and Data

```python
model = FeatureClassifier.from_model(modelLocation,data)
```

## Use Model to Predict Photo and Sort Photo Based on Prediction

```python
import shutil

person_dst = os.path.join(folder_dir,"Person")
notperson_dst = os.path.join(folder_dir,"Not Person")

for images in os.listdir(folder_dir):
    if images == "Not Person" or images == "Person":
        pass
    else:
        imageFilePath = os.path.join(folder_dir,images)
        results =  model.predict(imageFilePath, visualize=False, gradcam=False)
        categoryTensor = str(results[0]) 
        print(categoryTensor)
        if categoryTensor == "Person":
            print("Person")
            shutil.move(imageFilePath, person_dst)
        else:
            print("Not a Person")
            shutil.move(imageFilePath, notperson_dst)
```
